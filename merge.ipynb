```python
import pandas as pd
```

```python
expression=pd.read_csv(r'GeneExpression_with_SignificantPvalues.csv',  sep = ',')
disease_state =  pd.read_csv(r'PFC_Expression_with_DiseaseState.csv',  sep = ',')
```

```python
expression
```

```python
temp = disease_state.loc[0:1,:].copy()
temp.iloc[1,:]=temp.columns
temp = temp.T
temp_expr = expression.T.copy()
temp_expr['sample_title'] = temp_expr.index

df_merge = pd.merge(temp, temp_expr, right_on='sample_title', left_on=temp.columns[1], how="right")
df_merge.index = df_merge.iloc[:,0]
df_merge.columns = df_merge.iloc[1,:]
df_merge = df_merge.drop(columns=[df_merge.columns[0],df_merge.columns[2]])
df_merge = df_merge.iloc[2:232,:].T
df_merge.index.name = None
df_merge = df_merge.T
df_merge['Disease_State'] = df_merge['Disease_State'].str.strip()
#df_merge = df_merge.drop(columns=[df_merge.columns[0],df_merge.columns[1]])
#df_merge = df_merge.drop(['!Sample_title']).T
```

```python
Disease = df_merge.loc[df_merge['Disease_State'] == 'A']
Disease = Disease.T[1:]
Control = df_merge.loc[df_merge['Disease_State'] == 'N']
Control = Control.T[1:]
```

```python
Disease['Mean']=  Disease.mean(axis=1, skipna=True)
Control['Mean']=Control.mean(axis=1, skipna=True)
Disease['Ratio']=2**Disease['Mean']/2**Control['Mean']
```

```python
Control.index.name = 'ID'
Control.index = pd.to_numeric(Control.index)
```

```python
Deg = Disease.loc[(Disease["Ratio"] > 1)]
```

```python
Deg.index.name='ID'
Deg.index = pd.to_numeric(Deg.index)
deg_merge = pd.merge(Deg, Control, left_index=True, right_index = True,how="inner")
deg_merge.sort_index(axis=1)
```

```python
go_function=pd.read_csv('go_enrichment_from_annotation.txt', skiprows=11,sep="\t")
go_function.columns =['Function','Background_frequency','Sample_frequency','Expected_pvalue','fold_enrichment','over/underrepresentation','pvalue','FDR']
```

```python
go_function[['Function','go_ID']]=go_function['Function'].str.split('(', expand=True)
go_function['go_ID']=go_function["go_ID"].str[:-1]
```

```python
goID_list = go_function['go_ID'].tolist()
```

```python
annotation =  pd.read_csv(r'GPL4372_annotation.csv',  sep = ',')
annotation.columns = annotation.iloc[0] 
annotation = annotation.drop(annotation.index[[0, -1]]).reset_index(drop=True)
annotation.index = annotation['ID']
annotation.index = pd.to_numeric(annotation.index)
annotation=annotation.drop(columns='ID')
annotation
```

```python
go_deg =annotation[pd.DataFrame(annotation['GO:Function ID'].tolist()).isin(goID_list).any(1).values]
```

```python
go_deg
```

```python
final_deg = pd.merge(go_deg, deg_merge, left_on='ID', right_on='ID', how='inner')
expression_deg =final_deg.loc[:, final_deg.columns.str.contains('PFC')]
#final_deg =final_deg.rename(columns = lambda x : str(x)[:-4])
#final_deg[sorted(final_deg.columns)]
expression_deg.to_csv("DegExpression_after_Go_and_ttest.csv")
```

```python
annotation_deg= final_deg[final_deg.columns.drop(list(final_deg.filter(regex='PFC')))]
annotation_deg.to_csv("Annotation_after_Go_and_ttest.csv")
```
