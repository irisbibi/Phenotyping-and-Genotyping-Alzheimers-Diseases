```python
from tabulate import tabulate
import GEOparse
```

```python
gse = GEOparse.get_GEO(filepath="GSE44772_family.soft.gz")
```

```python
for gsm_name, gsm in gse.gsms.items():
    print("Name: ", gsm_name)
    print("Metadata:",)
    for key, value in gsm.metadata.items():
        print(" - %s : %s" % (key, ", ".join(value)))
    print ("Table data:",)
    print (gsm.table.head())
    break
```
